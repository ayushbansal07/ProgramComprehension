```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#utility function, gets the total gaze time for each line for 1 subject
def get_gaze_time(data):
    data = np.array(data)
    last_line = 0
    gaze = 0
    ans = []
    for line in data:
        if line[1] == 0:
            continue
        if line[1] != last_line:
            ans.append((last_line,gaze))
            last_line = line[1]
            gaze = line[5] - line[4] + 1
        else:
            gaze += line[5] - line[4] + 1
    ans.append((last_line,gaze))
    return ans[1:]
```

```python
#utility function, gets the total gaze time for IMPORTANT LINES ONLY for 1 subject
def gaze_time_imp(imp, data):
    data = np.array(data)
    total = 0
    for line in data:
        if line[1] == imp:
            total += line[5] - line[4] + 1
    return total
```

```python
import os
base_dir = "./data/datafiles_V3/"
#directory where all data files are stored. each data file is in a folder corresponding to code number.
#Example Structure of Directoy - 
#datafiles_V3   -> 5 -> abc5.xlsx, def5.xlsx ........
#               -> 24 -> abc24.xlsx, def24.xlsx ........
#               -> 32 -> abc32.xlsx, def32.xlsx ........
CODE_NUMS = [5,24,32] #All the code numbers
data = {5 : [], 24 : [], 32 : []}
for code in CODE_NUMS:
    for x in os.listdir(base_dir + str(code)):
        data[code].append(pd.read_excel(base_dir + str(code) + "/" + x))

#All data files for a code read in a dictonary for that code, i.e., data[5] contains all data files for code 5
```

```python
LINES = {5 : 21, 24 : 24, 32 : 35} #Number of Lines in each code

#just a helper function to initialise an empty map
def init_map(m,tp=0):
    #tp -> 0 - [] array, 1 - int (0)
    for code in CODE_NUMS:
        if tp == 0:
            m[code] = []
        elif tp == 1:
            m[code] = 0
    return m
```

```python
TOTAL_GAZE_TIMES = {}
#TOTAL_GAZE_TIMES contains total time spent by each subject on a code
TOTAL_GAZE_TIMES = init_map(TOTAL_GAZE_TIMES)
for code in CODE_NUMS:
    for sub in range(len(data[code])):
        myarr = np.array(data[code][sub])
        total = 0
        for line in myarr:
            if line[1] != 0:
                total += line[5] - line[4] + 1
        TOTAL_GAZE_TIMES[code].append(total)

TOTAL_GAZE_TIMES
```

```python
#List of Important Lines for each code
IMP_LINES = {5 : [6, 8, 10, 11, 13, 16], 24: [9, 10, 11], 32 : [8, 14, 24, 25, 26]}
```

```python
#create the output directories if the don't exist
if not os.path.exists("./plots"):
    os.makedirs("./plots")
if not os.path.exists("./plots/percent_fixation_all"):
    os.makedirs("./plots/percent_fixation_all")
if not os.path.exists("./plots/percent_fixation_imp"):
    os.makedirs("./plots/percent_fixation_imp")
```

```python
gaze_all_perc = {}
#contains percentage fixation time on each line for each subject
gaze_all_perc = init_map(gaze_all_perc)
for code in CODE_NUMS:
    for sub in range(len(data[code])):
        temp_perc = []
        for line in np.arange(LINES[code]):
            gaze_time = gaze_time_imp(line + 1,data[code][sub])
            temp_perc.append(gaze_time/TOTAL_GAZE_TIMES[code][sub])
        gaze_all_perc[code].append(temp_perc)
```

```python
#Plotting percentage fixation time on each line for each subject
font = {'size'   : 24}
plt.rc('font', **font)
PLOT_BLACKLISTED = {5 : [], 24: [], 32 : []} #The list of subjects that are not to be plotted, ignore if all subjects have to be plotted.
FIGSIZES = {5 : (30,15), 24 : (20,10), 32 : (30,15)} #Adjust this for the size of the graph to make it of appropriate size
for code in CODE_NUMS:
    plt.figure(figsize=FIGSIZES[code])
    num_subs = len(data[code]) - len(PLOT_BLACKLISTED[code])
    ct = 0
    for sub in range(len(data[code])):
        if sub in PLOT_BLACKLISTED[code]:
            continue
        if ct == int((num_subs-1)/2):
            plt.bar(np.arange(LINES[code])+ct*(0.8/(num_subs + 1)),np.array(gaze_all_perc[code][sub]),width=(0.8/(num_subs + 1)),tick_label=np.arange(LINES[code])+1)
        else:
            plt.bar(np.arange(LINES[code])+ct*(0.8/(num_subs + 1)),np.array(gaze_all_perc[code][sub]),width=(0.8/(num_subs + 1)))
        ct += 1
    #plt.show()
    plt.xlabel("Line Number")
    plt.ylabel("Fraction of total fixation time")
    plt.savefig("./plots/percent_fixation_all/code_{}.png".format(code))
```

```python
gaze_imp_perc = {}
#contains percentage fixation time for ONLY IMPORTANT LINES for each subject
gaze_imp_perc = init_map(gaze_imp_perc)
for code in CODE_NUMS:
    for sub in range(len(data[code])):
        temp_perc = []
        for imp in IMP_LINES[code]:
            gaze_time = gaze_time_imp(imp,data[code][sub])
            temp_perc.append(gaze_time/TOTAL_GAZE_TIMES[code][sub])
        gaze_imp_perc[code].append(temp_perc)
        
```

```python
#Plotting percentage fixation time for ONLY IMPORTANT LINES for each subject
font = {'size'   : 24}
plt.rc('font', **font)
PLOT_BLACKLISTED = {5 : [], 24: [], 32 : []} #The list of subjects that are not to be plotted, ignore if all subjects have to be plotted.
FIGSIZES = {5 : (15,15), 24 : (10,10), 32 : (20,20)} #Adjust this for the size of the graph to make it of appropriate size
for code in CODE_NUMS:
    plt.figure(figsize=FIGSIZES[code])
    num_subs = len(data[code]) - len(PLOT_BLACKLISTED[code])
    ct = 0
    for sub in range(len(data[code])):
        if sub in PLOT_BLACKLISTED[code]:
            continue
        if ct == int((num_subs-1)/2):
            plt.bar(np.arange(len(IMP_LINES[code]))+ct*(0.8/(num_subs + 1)),np.array(gaze_imp_perc[code][sub]),width=(0.8/(num_subs + 1)),tick_label=IMP_LINES[code])
        else:
            plt.bar(np.arange(len(IMP_LINES[code]))+ct*(0.8/(num_subs + 1)),np.array(gaze_imp_perc[code][sub]),width=(0.8/(num_subs + 1)))
        ct += 1
    #plt.show()
    plt.xlabel("Line Number (Important Only)")
    plt.ylabel("Fraction of total fixation time")
    plt.savefig("./plots/percent_fixation_imp/code_{}.png".format(code))

```
