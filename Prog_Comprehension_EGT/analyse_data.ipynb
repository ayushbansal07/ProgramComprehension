```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("data/V3.xlsx")
data.head()
```

```python
data_np = np.array(data)
data_np
```

```python
srt_by_ct = sorted(data_np,key = lambda x:x[3],reverse=True)
```

```python
srt_by_ct = np.array(srt_by_ct)
for datal in srt_by_ct:
    if datal[2] != '\'blank\'':
        print(datal[2][1:-1],datal[3])
```

```python
#2,6,8
tokmap = {}
for d in data_np:
    temp = (d[2],d[6],d[8])
    if temp not in tokmap:
        tokmap[temp] = d[3]
    else:
        tokmap[temp] += d[3]

```

```python
for key, val in sorted(tokmap.items(), key = lambda x: (x[1],x[0]),reverse = True):
    print(key,val)
```

```python
srt_by_time = sorted(data_np,key = lambda x:x[5]-x[4],reverse=True)
```

```python
srt_by_time = np.array(srt_by_time)
for datal in srt_by_time:
    if datal[2] != '\'blank\'':
        print(datal[2][1:-1],datal[5]-datal[4])
```

```python
tokmap_time = {}
for d in data_np:
    temp = (d[2],d[6],d[8])
    if temp not in tokmap_time:
        tokmap_time[temp] = d[5] - d[4]
    else:
        tokmap_time[temp] += d[5] - d[4]
        

```

```python
for key, val in sorted(tokmap_time.items(), key = lambda x: (x[1],x[0]),reverse = True):
    print(key,val)
```

```python
tok_name = {}
tok_stmt = {}
tok_block = {}
tok_func = {}
for d in data_np:
    if d[2] not in tok_name:
        tok_name[d[2]] = d[5] - d[4]
    else:
        tok_name[d[2]] += d[5] - d[4]
    
    if d[6] not in tok_stmt:
        tok_stmt[d[6]] = d[5] - d[4]
    else:
        tok_stmt[d[6]] += d[5] - d[4]
    
    if d[8] not in tok_block:
        tok_block[d[8]] = d[5] - d[4]
    else:
        tok_block[d[8]] += d[5] - d[4]
        
    if d[11] not in tok_func:
        tok_func[d[11]] = d[5] - d[4]
    else:
        tok_func[d[11]] += d[5] - d[4]
    
```

```python
print("By token_name\n")
tokname = []
tokct = []
for key, val in sorted(tok_name.items(), key = lambda x: (x[1],x[0]),reverse = True):
    print(key,val)
    if val > 0:
        tokname.append(key)
        tokct.append(val)
```

```python
print("By stmt_name\n")
stmtname = []
stmtct = []
for key, val in sorted(tok_stmt.items(), key = lambda x: (x[1],x[0]),reverse = True):
    print(key,val)
    stmtname.append(key)
    stmtct.append(val)
```

```python
print("By block_name\n")
blockname = []
blockct = []
for key, val in sorted(tok_block.items(), key = lambda x: (x[1],x[0]),reverse = True):
    print(key,val)
    blockname.append(key)
    blockct.append(val)
```

```python
print("By function_name\n")
funcname = []
funcct = []
for key, val in sorted(tok_func.items(), key = lambda x: (x[1],x[0]),reverse = True):
    print(key,val)
    funcname.append(key)
    funcct.append(val)
```
