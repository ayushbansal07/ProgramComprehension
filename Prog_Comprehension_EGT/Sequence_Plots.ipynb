```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_seq_map(data):
    data = np.array(data)
    counts = {}
    last_line = 0
    total_lines = 0
    for line in data:
        if line[0]==0:
            continue
        if line[1] != last_line:
            total_lines += 1
        if line[1] == last_line+1:
            if line[1] not in counts:
                counts[line[1]] = 0
            counts[line[1]] += 1
        last_line = line[1]
    return counts, total_lines
```

```python
import os
base_dir = "./data/datafiles_V3/"
#directory where all data files are stored. each data file is in a folder corresponding to code number.
#Example Structure of Directoy - 
#datafiles_V3   -> 5 -> abc5.xlsx, def5.xlsx ........
#               -> 24 -> abc24.xlsx, def24.xlsx ........
#               -> 32 -> abc32.xlsx, def32.xlsx ........
CODE_NUMS = [5,24,32] #All the code numbers
data = {5 : [], 24 : [], 32 : []}
for code in CODE_NUMS:
    for x in os.listdir(base_dir + str(code)):
        data[code].append(pd.read_excel(base_dir + str(code) + "/" + x))

#All data files for a code read in a dictonary for that code, i.e., data[5] contains all data files for code 5
```

```python
LINES = {5 : 21, 24 : 24, 32 : 35} #Number of Lines in each code

#just a helper function to initialise an empty map
def init_map(m,tp=0):
    #tp -> 0 - [] array, 1 - int (0)
    for code in CODE_NUMS:
        if tp == 0:
            m[code] = []
        elif tp == 1:
            m[code] = 0
    return m
```

```python
#create the output directories if the don't exist
if not os.path.exists("./plots"):
    os.makedirs("./plots")
if not os.path.exists("./plots/sequence_analysis"):
    os.makedirs("./plots/sequence_analysis")
```

```python
sequence_maps = {}
#Sequence Maps contain the number of lines in sequence (from line i to line i+1)
sequence_maps = init_map(sequence_maps)
for code in CODE_NUMS:
    for sub in range(len(data[code])):
        sequence_maps[code].append(get_seq_map(data[code][sub]))
```

```python
tok_analysis = {}
#Token Analysis contain the percentage lines in sequence (from line i to line i+1)
tok_analysis = init_map(tok_analysis)

for code in CODE_NUMS:
    for m, total in sequence_maps[code]:
        line_ct = np.zeros(LINES[code])
        for k,v in m.items():
            line_ct[k] = v/total
        tok_analysis[code].append(line_ct)
```

```python
#Plotting percentage lines in sequence
font = {'size'   : 24}
plt.rc('font', **font)
PLOT_BLACKLISTED = {5 : [], 24: [], 32 : []}  #The list of subjects that are not to be plotted, ignore if all subjects have to be plotted.
FIGSIZES = {5 : (30,15), 24 : (20,10), 32 : (30,15)} #Adjust this for the size of the graph to make it of appropriate size
plt.show()
for code in CODE_NUMS:
    plt.figure(figsize=FIGSIZES[code])
    num_subs = len(data[code]) - len(PLOT_BLACKLISTED[code])
    ct = 0
    for sub in range(len(data[code])):
        if sub in PLOT_BLACKLISTED[code]:
            continue
        #plt.bar(np.arange(LINES[code])+sub*0.2,armstrong_tok_analysis[sub],width=0.2)
        if ct == int((num_subs-1)/2):
            plt.bar(np.arange(LINES[code])+ct*(0.8/(num_subs + 1)),np.array(tok_analysis[code][sub]),width=(0.8/(num_subs + 1)),tick_label=1+np.arange(LINES[code]))
        else:
            plt.bar(np.arange(LINES[code])+ct*(0.8/(num_subs + 1)),np.array(tok_analysis[code][sub]),width=(0.8/(num_subs + 1)))
        ct += 1
    #plt.show()
    plt.xlabel("Line Number")
    plt.ylabel("Fraction of switches in sequence from line i to i+1")
    plt.savefig("./plots/sequence_analysis/code_{}.png".format(code)) 
```
