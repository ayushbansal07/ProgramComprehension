```python
import pandas as pd
```

```python
data = pd.read_excel('k_20.xlsx')
```

```python
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
filtered_data = np.array(data)
```

```python
filtered_data
```

```python
fil_data = []
for x in filtered_data:
    if x[1] !=0:
        fil_data.append((x[1],x[-1]))
```

```python
filtered_data = np.array(fil_data)
```

```python
filtered_data
```

```python
plt.figure(figsize=(20,12))
plt.scatter(filtered_data[:,1],filtered_data[:,0])
plt.ylim(60,0)
plt.savefig('myplot.png')
```

```python
mwa = 0
mwa_data = []
for x in filtered_data:
    mwa = 0.8*mwa + 0.2*x[0]
    mwa_data.append((mwa,x[1]))
```

```python
mwa_data = np.array(mwa_data)
```

```python
mwa_data
```

```python
plt.figure(figsize=(20,12))
plt.scatter(mwa_data[:,1],mwa_data[:,0])
plt.ylim(60,0)
plt.savefig('myplot_mwa.png')
```

```python
def get_going_back_time(data):
    max_line = np.max(data[:,0])
    print(max_line)
    ln = len(data)
    for i in range(1,ln):
        for j in range(10):
            if i-1-j >=0 and -data[i,0] + data[i-1-j,0] >= 0.5*max_line:
                print(data[i], data[i-1-j])
```

```python
get_going_back_time(filtered_data)
```
